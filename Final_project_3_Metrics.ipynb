## Первый вопрос

Поскольку нас просят оценить результаты прошедших событий (любых, я так понимаю - в задании не конкретизировано каких именно)
<br>
Поэтому обозначим все возможные метрики, которые могут помочь обрисовать некую картину игроков и их действий во время игр

### 1. Винрейт на каждом из уровней


Данная метрика позволяет понять, **насколько сложные уровни в игре**. 
<br>
Считается как - *Отношение кол-ва игроков, начавших уровень к кол-ву игроков, закончивших его*

Если винрейт **низок**, значит *уровень* является *сложным* для игроков.
<br>
Это может привести к *отвалу игроков*, т.к. они не смогут выполнить тематические задания для получения наград!

###  2. Процент по полученным предметам Во время тематического события и После него

Данная метрика позволит понять, не сломалась ли **вероятность дропа хороших плюшек** игрокам.
<br>
Например, в известной серии FIFA - режим Ultimate Team часто проводятся разного рода события, посвященные какой-то тематике в футболе. И при выполнении определенных заданий игроку даются **паки** с увеличенным на N процентов возможностью получения лучших футболистов в свою команду.
<br>
<br>
Так и здесь - нужно отслеживать сиутацию, при которой у нас бы не получилось так, что крутые ингейм предметы выпадали бы слишком часто, ДАЖЕ при условии, что у нас идет тематическое событие.
<br>
Считается как - *Отношение кол-ва упавших ингейм бонусов во время события к кол-ву упавших ингейм бонусов до события*

### 3.  Вовлеченность игроков в событие

Данная метрика показывает, насколько игроки вообще заинтересованы в данном событии (какой процент об активных игроков заинтересовался этим событием)
<br>
Считаеся как - *Отношение кол-ва игроков, участвующих в данном событии к Общему числу игроков за N-ый период*

### 4. ARPU/ARPPU (на событии)

Проверим, есть ли взаимосвязь между тем, сколько денег выручки приходилось в среднем на игрока
<br>
Вполне возможна такая картина, что тематическое событие подстегнет игроков к приобретению внутриигровых предметов
<br>
Считается как - *Выручка от игроков, участвующих в событии к Кол-ву игроков, участвующих в событии*

### 5. Конверсия 

Данная метрика позволит проверить предположение - изменилась ли конверсия в покупку игроков за время события?
<br>
Смогло ли оно стимулировать увеличение выручки?

Считаеся как - *Отношение платящих игроков к Общему числу игроков (за период N)*

### 6. Относительная конверсия

Метрика отражает *Отношение числа игроков, которые выполнили какое-либо целевое действие* (прохождение уровней, в данной ситуации) *к Общему числу игроков, дошедших до предшествующего уровня*

Таким образом, мы можем сформировать понимание того, как формируется воронка из игроков во время нашего тематического события

### 7. DAU/MAU + ASL

Метрика позволит ответить на вопрос - изменился ли паттерн поведения игроков за время тематического события?

Вполне возможно, что игроки, стимулируемые наградами за прохождение уровней во время ивента, будут и в дальнейшем 
чаще заходить в игру и проводить там время

### 8. Retention rate 

По логике с предыдущим пунктом - тематическое событие может увеличить RR игроков (даже увеличить его!)
<br>
<br>
Считается как - *Кол-во игроков на конец периода (тематическое событие) - кол-во новых игроков за период / Кол-во игроков до начала события (в начале периода)*

### 9. NPS 

Данная метрика поможет узнать, изменилась ли лояльность пользователей с проведением такого рода событий
<br>
Считается как - *Кол-во игроков, учавствоваших в предыдущем тематическом событии к Кол-ву игроков, участвующх в нынешнем событии*

Так мы поймем, помог ли наш ивент продвинуться в кол-ве лояльных игроков, или нет

### 10. Фактор виральности 

Можем выяснить, является ли наш ивент виральным событием - может ли он привлечь новых игроков
<br>
сугубо исходя из тематики ивента?


Например, Fornite - часто идут какие-либо коллаборации с брендами, актерами и персонажами, что позволяет привелчь 
новую аудиторию
